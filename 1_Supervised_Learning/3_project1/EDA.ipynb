## Data Cleaning

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122, "Region"] = 0
```

```python
dataset.info()
```

```python
dataset[["Region"]] = dataset[['Region']].astype('int32')
```

```python
dataset.isnull().sum()
```

```python
## Remove the null values
df = dataset.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

* **Changes the requried columns as integer data type** 

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

#### changing the others columns to float datatype

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
## Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(11, 8))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel("Number of fires", weight='bold', fontsize=25)
plt.xlabel('Months', weight='bold', fontsize=25)
plt.title('Fire Analysis of Bejaia Region', weight='bold', fontsize=25)
plt.show()
```
