### **Decision Tree Implementation**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
# independent features
X = pd.DataFrame(iris['data'], columns=['sepal length in cm', 'sepal width in cm', 'petal lenght in cm', 'petal width in cm'])
X.head()
```

```python
# dependent features
y = iris['target']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

* In `Decision Tree` we need not to apply **standrazition** or any kind of **normalization**

```python
## apply decision tree classifier 
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier()
```

```python
treeClassifier.fit(X_train, y_train)
```

```python
## Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(16, 15))
tree.plot_tree(treeClassifier, filled=True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred = treeClassifier.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

---
`Preventing overfiting:` 
1. **Post Pronning**:- first create whole tree then pronne it. (good for smaller dataset)
2. **Pre Pronning**

- here in the tree we can clearly see that after depth=2 we are getting a proper splitting

```python
treeModel = DecisionTreeClassifier(max_depth=2)
```

```python
treeModel.fit(X_train, y_train)
```

```python
## Visualize the Decision Tree
plt.figure(figsize=(10, 8))
tree.plot_tree(treeModel, filled=True)
plt.show()
```

```python
y_pred = treeModel.predict(X_test)
```

```python
y_pred
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
