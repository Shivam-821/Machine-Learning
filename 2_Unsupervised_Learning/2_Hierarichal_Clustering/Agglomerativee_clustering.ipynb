### Agglomerative Clustering:-
- Recursively merges of clusters of sample data util only single cluster left

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets
```

```python
## load iris dataset
iris = datasets.load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_data.head()
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
## Apply PCA to decrease the dimension from 4 to 2
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(X_scaled)
pca_scaled
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c=iris.target)
```

```python
## Agglomerative Clustering
# step 1: plot dendrogram to find value of k(centroid center).

import scipy.cluster.hierarchy as sc
#plot the dendrogram
plt.figure(figsize=(20, 8))

## create dendrogram
sc.dendrogram(sc.linkage(pca_scaled, method='ward'))
plt.xlabel('Sample Index')
plt.ylabel("Eucledian Distance")
plt.title('DendroGram')
```

- It is clearly visiable that largest vertical line which is not cut by any horizontal line is blue one(smaller one): 2 points

**k = 2**

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c=cluster.labels_)
```
